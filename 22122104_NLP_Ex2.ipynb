# Handling PDFs in Python: Reading, Creating, Inserting Content



```python
pip install PyPDF2
```

```python
from PyPDF2 import PdfReader
import PyPDF2
```

```python
#creating a pdf reader object
reader=PdfReader('BasePaper')
```

```python
print(len(reader.pages))
```

```python
page=reader.pages[0]
```

```python
#extracting text from page
text=page.extract_text()
print(text)
```

## Create and Insert pages into PDF file

```python
pdf_writer=PyPDF2.PdfWriter()
```

```python
pdf_writer.add_page(page) #added in buffer
```

```python
pdf_output=open("BaseExtraction.pdf","wb") #created a new pdf file for the output
```

```python
pdf_writer.write(pdf_output) #shifted to the new pdf file created
```

```python
pdf_output.close()
```

## Merging PDF Files

For merging pages extracted in different pdfs into a single one


```python
from PyPDF2 import PdfMerger
```

```python
#Create an instance of PdfMerger() class
merger=PdfMerger()
```

```python
#Create a list with file paths
pdf_file=['BasePaper', 'BaseExtraction.pdf']
```

```python
for pdf_file in pdf_file:
  #Append PDF files
  merger.append(pdf_file)
```

```python
#Write out the merged PDF file
merger.write("mergedpdf.pdf")
merger.close()
```
